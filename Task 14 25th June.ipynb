Question 1:
<br>Replace null values 'Age' column with ffill()
<br>Replace null values in 'Cabin and 'Embarked' columns with most occured value

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(method = "ffill", inplace = True)
titanic
```

```python
titanic['Cabin'].fillna(titanic['Cabin'].mode().iloc[0], inplace = True)
titanic['Embarked'].fillna(titanic['Embarked'].mode().iloc[0], inplace = True)
titanic
```

```python
titanic.isnull().sum()
```

Question 2:
<br>After handling missing values, Convert Cabin, Embarked columns into numerical columns

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
titanic['Cabin'] = le.fit_transform(titanic['Cabin'])
titanic['Embarked'] = le.fit_transform(titanic['Embarked'])
titanic['Name'] = le.fit_transform(titanic['Name'])
titanic['Sex'] = le.fit_transform(titanic['Sex'])
titanic['Ticket'] = le.fit_transform(titanic['Ticket'])
titanic
```
